# While Loops

```python
number = 0
while number < 8:
    print(number)
    number = number + 1
```

```python
number = 0
while number < 8:
    print(number)
    if number == 5:
        break
    number = number + 1
```

```python
number = 0
while number < 8:
    print(number)
    if number == 9:
        break
    number = number + 1
else:
    print('N/A < 8')
```

```python
number = 0
while number < 8:
    print(number)
    if number == 5:
        continue
    number = number + 1
else:
    print('N/A < 8')
```

```python
number = 0

while number < 8:
    number = number + 1
    if number == 6:
        continue
    print(number)
else:
    print('N/A < 8')
```
